```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torchvision
from neuralprophet import NeuralProphet

import os
for dirname, _, filenames in os.walk('/content/drive/MyDrive/Colab Notebooks'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Samsung Electronics.csv")
```

```python
df.rename(columns = {'Date' : 'ds', 'Close' : 'y'}, inplace = True)
```

```python
df
```

```python
df['ds'] = pd.to_datetime(df['ds'])
```

```python
model = NeuralProphet()
model.fit(df[['ds', 'y']], freq = 'D', epochs = 150)
```

```python
future = model.make_future_dataframe(df[['ds', 'y']], periods = 730)
```

```python
future
```

```python
forecast = model.predict(future)
```

```python
present = model.predict(df[['ds', 'y']])
```
